```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

from sklearn.preprocessing import StandardScaler, LabelEncoder
import csv
import os
import pickle

import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('test.csv', index_col="id")
df.head()
```

```python
df.shape
```

```python
numerical_features = [feature for feature in df.columns if df[feature].dtype != 'O' and feature != "y"]
categorical_features = [feature for feature in df.columns if df[feature].dtype == "O" and feature != "y"]
numerical_features, categorical_features
```

```python
scaler_path = "preprocessing/scaler.pkl"
label_encoder_path = "preprocessing/label_encoder.pkl"

def load_preprocessing_tools(tool_path):
    with open(tool_path, 'rb') as file:
        tool = pickle.load(file)
        return tool
    
scaler = load_preprocessing_tools(scaler_path)
label_encoder = load_preprocessing_tools(label_encoder_path)

# Scaling Standard scaler 
scaler.fit(df[numerical_features])

# label encoding
for feature in categorical_features:
    df[feature] = label_encoder.fit_transform(df[feature])
            
```

```python
def evaluate_model(model_path, result_file_name='result.csv'):
    with open(model_path, 'rb') as file:
        loaded_model = pickle.load(file)
        
    y_pred = loaded_model.predict(df)

    dummy_df = df.copy()
    dummy_df['y'] = y_pred
    dummy_df['y'].to_csv(result_file_name, index=True)
```

```python
model_path = 'model/bbc_OverSampled.pkl'
result_file_name = 'results/bbc_OverSampled.csv'

evaluate_model(model_path, result_file_name)
```
